## The Task 

Write a function called `plot_PCA` which takes the following parameters:

1. `X`: features matrix
2. `y`: the target array
3. `n_components` : number of PCA components

The function should

* Return PCA matrix with given number of components
* Plot 3D representation of first 2 directs of the PCA Matrix with
    * Plot title as "First two PCA directions"
    * X label as "1st eigenvector"
    * Y label as "2nd eigenvector"

Here are the libraries you can import and use:

```python

```

## Solution 

````python
from sklearn.decomposition import PCA 
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

def plot_PCA(X, y, n_components=2):
    
    try:
        n_components = int(n_components)
    except ValueError:
        print("Please ensure the value of n_components is a Python integer.", "You entered: ", n_components)
    
    model = PCA(n_components=n_components)            
    model.fit(X)                           
    X_PCA = model.transform(X)
    
    fig = plt.figure(1, figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=110)
    ax.scatter(X_PCA[:, 0], X_PCA[:, 1], c=y, cmap=plt.cm.Paired)
    ax.set_title("First two PCA directions")
    ax.set_xlabel("1st eigenvector")
    ax.w_xaxis.set_ticklabels([])
    ax.set_ylabel("2nd eigenvector")
    ax.w_yaxis.set_ticklabels([])
    plt.show()
    
    return X_PCA
````
